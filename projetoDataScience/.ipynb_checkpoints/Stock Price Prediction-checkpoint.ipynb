# Importando as bibliotecas 

```python
import pandas as pd
import yfinance as yf
import datetime
from datetime import date, timedelta
from sklearn.model_selection import train_test_split
```

# Setando as Datas 

```python
#pegando a data de hoje
hoje= date.today()
dia_fim = hoje.strftime("%Y-%m-%d")

# Vamos 500 dias atras a parti de hoje
dia = date.today() - timedelta(days=5000)

dia_inicio = dia.strftime("%Y-%m-%d")

print(dia_inicio)
print(dia_fim)
```

# Pegando os dados do DataFrame 
             

```python
# Vamos analisar Petrobras 

df = yf.download('PBR', start=dia_inicio,  end=dia_fim, progress=False)
df.head()
```

# Dando um tapa no visual do DataFrame
## Adicionando um indice

Vamos criar indice, que seja valor inteiro 1,2,3, ...

```python
# estamos criando index na coluna Date, repare que vamos ter duas colunas 
df["Date"] = df.index
df
```

A função reset_index() é utilizada para redefinir o índice de um DataFrame no pandas, que é uma biblioteca muito usada para análise de dados em Python. O método reset_index() tem dois parâmetros principais que você mencionou: drop e inplace.

    drop=True: Este parâmetro indica se a coluna que anteriormente continha o índice deve ser descartada após a redefinição. Se drop for definido como True, a coluna do índice será removida do DataFrame resultante. Se for definido como False, a coluna do índice será mantida como uma coluna regular no DataFrame.

    inplace=True: Se definido como True, a redefinição do índice será feita diretamente no próprio DataFrame, ou seja, o DataFrame original será modificado. Se for definido como False, um novo DataFrame será retornado com o índice redefinido e o DataFrame original permanecerá inalterado.

```python
df.reset_index(drop=True, inplace=True)
df
```

```python
#Organizando a ordem, so colocando Date em primeiro 

df = df[["Date", "Open", "High", "Low", "Close", "Adj Close", "Volume"]]
df.head()    
```
